Para este Analasis Exploratorio de Datos contaremos con los datos completos luego de haber realizado el merge de los 3 dataframes presentes en el ETL (games_and_reviews.csv) y también con el dataframe acotado a 3000 filas que se utilizó para el deploy de la api a travez de render.com (games_and_reviews_sample.csv) 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_completa=pd.read_csv("../Dataset_ETL/games_and_reviews.csv")
data_acotada=pd.read_csv("../games_and_reviews_sample.csv")
```

ANALISIS DE CORRELACIONES

```python
#para el dataframe completo:
#seleccionamos columnas numéricas para la correlación
#nos aseguramso de ajustar estas columnas según tu dataset
columnas_numericas = ['playtime_forever', 'sentiment_score']
df_numeric = data_completa[columnas_numericas]

# Calculando la matriz de correlación
corr_matrix = df_numeric.corr()

# Visualizando la matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

```python
#para el dataframe acotado:
#seleccionamos columnas numéricas para la correlación
#nos aseguramso de ajustar estas columnas según tu dataset
columnas_numericas = ['playtime_forever', 'sentiment_score']
df_numeric = data_acotada[columnas_numericas]

# Calculando la matriz de correlación
corr_matrix = df_numeric.corr()

# Visualizando la matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

ANALISIS DE TENDENCIAS Y PATRONES

```python
#con el dataframe completo
#visualizamos de tendencias del tiempo de juego a lo largo de los años
plt.figure(figsize=(12, 6))
sns.lineplot(data=data_completa, x='posted_year', y='playtime_forever', estimator='mean', ci=None)
plt.title('Tendencia del Tiempo de Juego por Año')
plt.xlabel('Año')
plt.ylabel('Tiempo de Juego Promedio (horas)')
plt.show()
```

```python
#con el dataframe acotado
#visualizamos de tendencias del tiempo de juego a lo largo de los años
plt.figure(figsize=(12, 6))
sns.lineplot(data=data_acotada, x='posted_year', y='playtime_forever', estimator='mean', ci=None)
plt.title('Tendencia del Tiempo de Juego por Año')
plt.xlabel('Año')
plt.ylabel('Tiempo de Juego Promedio (horas)')
plt.show()
```
