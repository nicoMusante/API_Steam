```python
import pandas as pd
import json
```

```python
ruta="../Dataset_Original/australian_users_items.json"
```

```python
with open(ruta, 'r', encoding='utf-8') as w:
    data = w.readlines() #la funcion eval se utiliza para evaluar expresiones de cadena que no son seguras, esto se hace así debido a que no es un json puro el json que estamos manejando  
    data = [eval(line.strip()) for line in data]
df=pd.DataFrame(data)
```

```python
df #exploramos el dataframe
```

```python
df=df.drop(["items_count", "steam_id", "user_url"], axis=1) #borramos las columnas que no vamos a utilizar
df
```

```python
filas_con_nulos = df[df.isnull().any(axis=1)] #verificamos la existencia de nulos
filas_con_nulos #no existen nulos en el dataframe
```

```python
#eliminamos todos los duplicados del dataframe
df=df.drop_duplicates(subset="user_id")
```

```python
#desanidamos la columna items en otro dataframe para mas adelante utilizarla en la api de forma más práctica
df_items = pd.json_normalize(data, record_path=['items'], meta=['items_count','user_id', 'user_url'] )
#data...usar su variable en donde guardaron cada linea al abrir el archivo
df_items
```

```python
df_items
```

```python
#eliminamos los duplicados de este nuevo dataframe
df_items.drop_duplicates(subset=["user_id", "item_name"])
```

```python
#finalmente guardamos ambos dataframe en 2 csv
df_items.to_csv("user_items.csv")
```
