```python
import pandas as pd
import json
```

```python
ruta="../Dataset_Original/output_steam_games.json"
```

```python
#a diferencia de los otros json este es un json puro por lo que podemos abrirlo normalmente
data = []
with open(ruta,'rt', encoding='utf-8') as f:
    for line in f:
        data.append(json.loads(line))
df = pd.DataFrame(data)
```

```python
#exploramos el dataframe
df
```

```python
#borramos las columnas que no vayamos a utilizar 
#en este caso como son varias tomamos directamente solo aquellas que usaremos
df=df[["id", "title", "developer", "genres"]]
```

```python
#borramos los nulos 
df=df.dropna()
df
```

```python
#borramos los duplicados
df=df.drop_duplicates(subset=["title"])
df
```

```python
# Verificamos y reemplazamos los valores que no son listas con listas vacías
df['genres'] = df['genres'].apply(lambda x: x if isinstance(x, list) else [])

# Aplicamos get_dummies a lo largo de las columnas (axis=1) y expandir a nuevas columnas
df_dummies = df['genres'].apply(lambda x: pd.Series({genre: 1 for genre in x}))

# Llenamos NaN en df_dummies con 0
df_dummies = df_dummies.fillna(0, downcast='int')

# Concatenamos el DataFrame original con las nuevas columnas dummy
dfgames = pd.concat([df, df_dummies], axis=1)

# Eliminamos la columna original "genres"
dfgames = dfgames.drop('genres', axis=1)
```

```python
#finalmente guardamos el dataframe en un csv
dfgames.to_csv("steam_games.csv")
```
