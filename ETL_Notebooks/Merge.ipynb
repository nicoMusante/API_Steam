```python
import pandas as pd
```

```python
reviews=pd.read_csv("reviews.csv")
items=pd.read_csv("user_items.csv")
games=pd.read_csv("steam_games.csv")
```

```python
#borramos en todos los dataframes la columna "Unnamed: 0" que guarda los indices del dataframe en el csv
games.drop(columns="Unnamed: 0", inplace=True)
reviews.drop(columns="Unnamed: 0", inplace=True)
items.drop(columns="Unnamed: 0", inplace=True)
```

```python
#ahora hacemos el merge de los 3 dataframes mediante los id de los items 
dfMaestro=pd.merge(games, items, left_on="id", right_on="item_id") 
#borramos la columna "items_id" ya que es la misma columna que "id" solo que con otro nombre
dfMaestro.drop("id", axis=1, inplace=True)
```

```python
#unimos el dataframe que posee los datos de "games" y de "items" con el de reviews 
dfMaestro=pd.merge(dfMaestro, reviews, on="user_id")
```

```python
#borramos aquellas columnas que se repitan o las que no necesitemos, 
# para esto oprimero vemos el resultado que nos quedo luego de hacer el merge
pd.set_option('display.max_columns', None)
dfMaestro
```

```python
dfMaestro.drop(["item_name", "playtime_2weeks", "items_count", "user_url", "item_id_y"], axis=1, inplace=True)
```

```python
dfMaestro.rename({"item_id_x": "item_id"}, axis=1, inplace=True)
```

```python
#guardamos el dataframe completo para luego utilizarlo en el EDA
dfMaestro.to_csv("games_and_reviews.csv")
```

```python
#reducimos las informaciÃ³n del dataframe para que sea mÃ¡s facil de manipular luego cuando tengamos que hacer
# el merge entre los dataframes, debido a que se necesita mas memoria ram de la que poseo

#de esta forma borramos todos los juegos que hayan sido jugados 0 horas o mas de 100 horas 

dfMaestro=dfMaestro.drop(dfMaestro[dfMaestro["playtime_forever"]>100].index) 
dfMaestro=dfMaestro.drop(dfMaestro[dfMaestro["playtime_forever"]==0].index) 
dfMaestro=dfMaestro.reset_index(drop=True)
```

```python
#finalmente guardamos reducimos el dataframe en el cual hicimos el merge para que a la hora de hacer el deploy
# de nuestra API no tengamos ningun inconveniente y lo guardamos un archivo csv para luego poder utlizarlo
dfMaestro=dfMaestro.sample(n=3000, random_state=42)
dfMaestro.to_csv("games_and_reviews_sample.csv")
```
